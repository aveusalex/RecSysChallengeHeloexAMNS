```python
import numpy as np
```

```python
# criando uma função que recebe uma lista e retorna os índices dos 5 maiores
def get_best_worst_indices(lista):
    # Converte a lista para um array NumPy
    arr = np.array(lista)

    if arr.shape[0] < 10:
        # Obtém os índices dos N maiores valores usando argsort
        indices_n_maiores = np.argsort(-arr)[:2]
        
    else:
        # Obtém os índices dos N maiores valores usando argsort
        indices_n_maiores = np.argsort(-arr)[:5]
       
    return indices_n_maiores.tolist()
```

```python
# aplicando na coluna stars
df_review['stars'] = df_review.stars.apply(get_best_worst_indices)
```

```python
# filtrando a coluna user_id para pegar apenas as reviews que estão nos indices de stars
df_review['users_embs'] = df_review.apply(lambda x: [x['user_id'][i] for i in x['stars']],axis=1)
```

```python
user_embeddings_dict = df_user.iloc[:, 2:].to_numpy().tolist()
user_embeddings_dict = {index: embedding for index, embedding in enumerate(user_embeddings_dict)}
df_user = df_user.set_index('user_id')

def get_user_embeddings(user_ids):
    embeddings = [user_embeddings_dict[user_id] for user_id in user_ids if user_id in user_embeddings_dict]
    if len(embeddings) == 0:
        return None
    return np.mean(embeddings, axis=0)

df_review['user_embeddings'] = df_review['users_embs'].apply(get_user_embeddings)
```

```python
# Crie um dicionário com os embeddings dos usuários
user_embeddings_dict = {}
for index, row in df_user.iterrows():
    user_embeddings_dict[index] = row.iloc[0:].to_numpy()

# Defina uma função que calcula a média dos embeddings dos usuários
def get_user_embeddings(user_ids):
    embeddings = []
    for user_id in user_ids:
        if user_id in user_embeddings_dict:
            embeddings.append(user_embeddings_dict[user_id])
    if len(embeddings) == 0:
        return None
    if len(embeddings) == 1:
        return embeddings[0]
    return np.mean(embeddings, axis=0)

# Crie uma nova coluna no dataframe de reviews com os embeddings dos usuários
df_review['user_embeddings'] = df_review['users_embs'].apply(get_user_embeddings)
```
